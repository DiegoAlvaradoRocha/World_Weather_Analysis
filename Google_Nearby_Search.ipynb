```python
# dependencies and set up
import requests
import gmaps

# import API key
from config import g_key
```

```python
# set the parameters for a hotel in Paris.
params = {
    "radius": 5000, 
    "type" : "lodging",
    "key": g_key, 
    "location" : "48.8566, 2.3522"
}
# Use the URL to search for hotels in Paris.
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

# make a request an get the JSON data from the search
hotels = requests.get(base_url, params = params).json()
hotels
```
